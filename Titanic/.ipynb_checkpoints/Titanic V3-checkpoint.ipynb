```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
print("Train: \n", df.isna().sum())
print("\n Test: \n", df_test.isna().sum())
```

```python
lst = [df, df_test]
for data in lst:
    data['Title'] = data['Name'].str.extract(r'([A-Za-z]*\.)')
    data['no_family'] = data['SibSp'] + data['Parch']

data
```

```python

for data in lst:
    print("Before: \n\n", data.isna().sum())
    data['Age'] = data['Age'].fillna(data['Age'].mean())
    data['Fare'] = data['Fare'].fillna(data['Fare'].mean())
    data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
    print("After: \n\n", data.isna().sum())
```

```python
for data in lst:
    data['has_cabin'] = 1
    data.loc[data['Cabin'].isna(),'has_cabin'] = 0
    
data
```

```python
df_test
```

```python
df['Title'] = df['Title'].map({'Mr.': 'Mr.', 'Miss.': 'Miss.', "Mrs.": "Mrs.", "Master.": 'Master.'})
df['Title'] = df['Title'].fillna('Others')

df_test['Title'] = df_test['Title'].map({'Mr.': 'Mr.', 'Miss.': 'Miss.', "Mrs.": "Mrs.", "Master.": 'Master.'})
df_test['Title'] = df_test['Title'].fillna('Others')
    
```

```python
df = df.drop(['PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Name'], axis=1)
df_test = df_test.drop(['PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Name'], axis=1)
print(df.head())
df_test
```

```python
df['Sex'] = df['Sex'].astype('category')
df['Title'] = df['Title'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')

df_test['Sex'] = df_test['Sex'].astype('category')
df_test['Title'] = df_test['Title'].astype('category')
df_test['Embarked'] = df_test['Embarked'].astype('category')
```

```python
df_response = df[["Survived"]]
df = df.drop('Survived', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_transformer

cat_features = ['Pclass', 'Sex', 'Embarked', 'Title', 'has_cabin']
num_features = ['Age', 'Fare', 'no_family']
x_tr, x_test, y_tr, y_test = train_test_split(df, df_response, test_size = 0.3, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_tr, y_tr, test_size = 0.3, random_state=42)


def feature_processor(x_train, x_val, df_test, x_test):
    preprocess = make_column_transformer((StandardScaler(), num_features),(OrdinalEncoder(), cat_features))
    preprocess.fit(x_train)
    train = preprocess.transform(x_train)
    val = preprocess.transform(x_val)
    dftest =  preprocess.transform(df_test)
    test = preprocess.transform(x_test)
    return train, val, dftest, test

def response_processor(y_train, y_val, y_test):
    encoder = LabelEncoder()
    encoder.fit(y_train)
    train = encoder.transform(y_train)
    val = encoder.transform(y_val)
    test = encoder.transform(y_test)
    return train, val, test

x_train, x_val, df_test_arr, x_test= feature_processor(x_train, x_val, df_test, x_test)
y_train, y_val, y_test = response_processor(y_train, y_val, y_test)


```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(8,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
early = EarlyStopping(monitor='val_loss', patience = 5)
modelcheckpoint = ModelCheckpoint('best_model.hdf5', save_best_only=True,monitor='val_loss')
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(np.asarray(x_train).astype('float32'), np.asarray(y_train).astype('float32'), batch_size=50, epochs=20, validation_data=(np.asarray(x_val).astype('float32'), np.asarray(y_val).astype('float32')), callbacks = [early, modelcheckpoint])
```

```python
from tensorflow.keras.models import load_model
best_model = load_model('best_model.hdf5')
```

```python
model.evaluate(x_test, y_test)
```

### Now to tune parameters to see if i can do better

```python
def get_model(activation, optimizer):
    model = Sequential()
    model.add(Dense(128, activation = activation, input_shape=(8,)))
    model.add(Dense(64, activation = activation))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])
    return model

from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
model = KerasClassifier(build_fn=get_model)

params={'activation':['relu', 'tanh'], 'optimizer':['adam', 'sgd'], 'epochs':[25,50,100], 'batch_size': [25,50,100]}

RS = RandomizedSearchCV(model, params, cv=5, random_state=42)
RS_results = RS.fit(np.asarray(x_train).astype('float32'), np.asarray(y_train).astype('float32'), validation_data=(np.asarray(x_val).astype('float32'), np.asarray(y_val).astype('float32')), callbacks = [early, modelcheckpoint])

    
    
              
```

```python
RS_results.best_params_
```

```python
#using best params
best = get_model('relu', 'adam')
best.fit(np.asarray(x_train).astype('float32'), np.asarray(y_train).astype('float32'), epochs=100, batch_size=100, validation_data=(np.asarray(x_val).astype('float32'), np.asarray(y_val).astype('float32')), callbacks = [early, modelcheckpoint])
best.evaluate(x_test, y_test) #doesnt seem to improve much
```
